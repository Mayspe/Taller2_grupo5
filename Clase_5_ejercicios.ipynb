Importar el dataset (está compartido en git) en Google colab.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/manquintana/taller_programacion_2_uade/master/Datasets/clase_5%20-%20netflix_titles.csv")
df
```

**¿Cuántas películas se lanzaron en 2019?:**

```python
peliculas_2019 = df[(df["type"] == "Movie") & (df["release_year"] == 2019) ]
print (f"En 2019 se lanzaron {len(peliculas_2019)} peliculas")
peliculas_2019
```

**¿Cuantas películas se lanzaron entre 2017 y 2018?:**

```python

lanzamientos =df[(df['release_year']>2016) & (df['release_year']<2019)] 
peliculas_2017_2018 = lanzamientos[lanzamientos['type'] == "Movie"]
print (f"Entre 2017 y 2018 se lanzaron {len(peliculas_2017_2018)} peliculas")
peliculas_2017_2018
```

**¿Qué cantidad de películas hay publicadas?**

```python
peliculas = df[df["type"] == "Movie"]
print (f"Hay {len(peliculas)} peliculas publicadas")
peliculas
```

**¿Cuántos shows de televisión hay publicados?:**

```python
tv_shows = df[df["type"] == "TV Show"]
print (f"Hay {len(tv_shows)} shows de televisión publicados")
tv_shows
```

Gráfico de barras con la cantidad de películas lanzadas por año

```python
peliculas = df[df["type"] == "Movie"]
peliculas

pelis_ordenadas =peliculas.groupby(["release_year"]).size().reset_index(name="cantidad")
pelis_ordenadas


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

barras=sns.catplot( x="release_year", y="cantidad", kind="bar", data= pelis_ordenadas)
```

```python
histograma= sns.relplot(x="release_year", y="cantidad", kind="line", data=pelis_ordenadas )
```

Gráficos con países de origen

```python
pelis_clean= peliculas.dropna()
pelis_clean = pelis_clean.groupby(["country"]).size().reset_index(name="cantidad_peliculas")
pelis_clean
```

```python
paises=pelis_clean["country"].str.split(",", n=1,  expand =True)
paises

pelis_clean["país"]= paises[0]
pelis_clean.drop(columns =["country"], inplace = True)

```

```python
import pycountry
def c_alpha_3(pais):
  try:
    return pycountry.countries.get(name=pais).alpha_3
  except:
    return None
pelis_clean["código"] = pelis_clean["país"].apply(lambda x :c_alpha_3(x))   
pelis_clean=pelis_clean.dropna()
#pelis_clean=pelis_clean.groupby(["código"]).()
pelis_clean

```

```python
pelis_clean= pelis_clean.groupby(["país", "código"], as_index=False).sum()
```

```python
import plotly.express as px
fig= px.choropleth(pelis_clean, locations="código", color="cantidad_peliculas", hover_name="país",  color_continuous_scale=px.colors.sequential.Pinkyl )
fig.show()
```
